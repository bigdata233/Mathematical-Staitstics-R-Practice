```r
#(1) 기초적 그래프 명령어
x<-c(1,3,5,7,9)   #정의역 x벡터에 지정
y<-c(2,3,4,1,2)   #정의역 y벡터에 지정
plot(x,y,main="picture name",xlab="x",ylab="y",xlim=c(0,10), ylim=c(0,4), type="l")   #선으로 그림을 그림
plot(x,y,main="picture name",xlab="x",ylab="y",xlim=c(0,10), ylim=c(0,4), type="h")   #수직선으로 그림을 그림

#(2) 확률밀도함수와 누적분포함수의 그림

#이산형 확률변수
#이항분포 B(5,0.2)의 그래프
x<-seq(0,5,1)   #정의역 생성
y<-dbinom(x, size=5, prob=0.2)    #pdf 값 구하기
plot(x,y,main='binomial distribution', type='h')

#포아송분포 Poisson(2)의 확률밀도함수
x<-seq(0,10,1)    #정의역 생성
y=dpois(x, lambda=2)    #pdf 값 구하기
plot(x,y,main="poisson distribution", type='h')

#연속형 확률변수
#균등분포 U(0,1)의 확률밀도함수와 누적분포함수
x<-seq(0,1,0.01)    #정의역 생성
y<-dunif(x,min=0,max=1)   #pdf 값 구하기
y2<-punif(x,min=0,max=1)   #cdf 값 구하기
plot(x,y,main="uniform distribution", type='l')
plot(x,y2,main="uniform cumulative distribution", type='l')

#베타분포 B(3,5)의 확률밀도함수와 누적분포함수
x<-seq(0,1,0.01)    #정의역 생성
y<-dbeta(x,shape1=3,shape=5)    #pdf 값 구하기
y2<-pbeta(x,shape1=3,shape=5)   #cdf 값 구하기
plot(x,y,main="Beta Distribution", type='l')
plot(x,y2,main='Beta Cumulative Distribution', type='l')

#t분포 t(7)의 확률밀도함수와 누적분포함수
x<-seq(-3,3,0.01)   #정의역 생성
y<-dt(x,df=7)   #pdf 값 구하기
y2<-pt(x,df=7)  #cdf 값 구하기
plot(x,y,main="Student-t Distribution", type='l')
plot(x,y2,main="Student-t Cumulative Distribution", type='l')

#카이제곱분포 chisq(5)의 확률밀도함수와 누적분포함수
x<-seq(0,20,0.01)    #정의역 생성
y<-dchisq(x,df=5)   #pdf 값 구하기
y2<-pchisq(x,df=5)  #cdf 값 구하기
plot(x,y,main="Chisquare Distribution", type='l')
plot(x,y2,main="Chisquare Cumulative Distribution", type='l')

#카이제곱분포 chisq(7;3)의 비중심 확률밀도함수와 누적분포함수
x<-seq(0,20,0.01)   #정의역 생성
y<-dchisq(x,df=7,ncp=3)   #pdf 값 구하기
y2<-pchisq(x,df=7,ncp=3)  #cdf 값 구하기
plot(x,y,main="noncentral parametric Chisquare Distribution", type='l')
plot(x,y2,main="noncentral parametric Chisquare Cumulative Distribution", type='l')

#F분포 F(3,5)의 확률밀도함수와 누적분포함수
x<-seq(0,7,0.01)    #정의역 생성
y<-df(x,df1=3,df2=5)    #pdf 값 구하기
y2<-pf(x,df1=3,df2=5)   #cdf 값 구하기
plot(x,y,main="F Distribution", type='l')
plot(x,y2,main="F Cumulative Distribution", type='l')

#F분포 F(3,5;3)의 비중심 확률밀도함수와 누적분포함수
x<-seq(0,7,0.01)    #정의역 생성
y<-df(x,df1=3,df2=5,ncp=3)    #pdf 값 구하기
y2<-pf(x,df1=3,df2=5,ncp=3)   #cdf 값 구하기
plot(x,y,main="noncentral parametric Chisquare Distribution", type='l')
plot(x,y2,main="noncentral parametric Chisquare Cumulative Distribution", type='l')
```
